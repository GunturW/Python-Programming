```python
import json
```

```python
headlineNews = []

with open('news.json', 'r') as f:
    line = f.readline()

    headline = json.loads(line)
    headlineNews.append(headline)
```

```python
headlineNews
```

```python
headlineNews1 = []

with open('news.json', 'r') as f:
    lines = f.readlines()

    for line in lines:
        headline = json.loads(line)
        headlineNews1.append(headline)
```
